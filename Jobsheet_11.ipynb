# Praktikum 1 Sequential Search pada Data Angka

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

# Praktikum 2 Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
      return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

# Praktikum 3 Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indeks {i}"
      return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

# Praktikum 4 Studi Kasus Nilai Mahasiswa dengan Binary Search

```python
def binary_search_nilai(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Nilai {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Nilai {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

# Praktikum 5 Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

# Praktikum 6 Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
  while j >= 0 and data[j] > key:
    data[j + 1] = data[j]
    j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
  if data[mid] == target:
    return f"{target} ditemukan di indeks {mid}"
  elif data[mid] < target:
    left = mid + 1
  else:
    right = mid - 1
  return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7] # ambil angka ke-8 sebagai target
print(binary_search(urut, target))
```

# Penugasan 4

```python
def binary_search_with_count(data, target):
    low = 0
    high = len(data) - 1
    langkah = 0

    while low <= high:
        langkah += 1
        mid = (low + high) // 2
        if data[mid] == target:
            return mid, langkah
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1, langkah

barang = [("Buku", 5000), ("Pena", 3000), ("Tas", 50000), ("Penggaris", 2000)]

barang.sort(key=lambda x: x[1])

list_harga = [b[1] for b in barang]

print("Daftar barang urut harga:", barang)
try:
    target_harga = int(input("Masukkan harga barang yang dicari: "))

    idx, step = binary_search_with_count(list_harga, target_harga)

    if idx != -1:
        nama_barang = barang[idx][0]
        print(f"Hasil: Barang '{nama_barang}' ditemukan dengan harga {target_harga}.")
        print(f"Jumlah langkah (iterasi): {step}")
    else:
        print(f"Harga {target_harga} tidak ditemukan dalam daftar.")
except ValueError:
    print("Error: Harap masukkan angka untuk harga.")
```

# Penugasan 5

```python
import time

def sequential_search_numeric(data, target):
    perbandingan = 0
    for i in range(len(data)):
        perbandingan += 1
        if data[i] == target:
            return i, perbandingan
    return -1, perbandingan

def binary_search_numeric(data, target):
    low = 0
    high = len(data) - 1
    langkah = 0
    while low <= high:
        langkah += 1
        mid = (low + high) // 2
        if data[mid] == target:
            return mid, langkah
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1, langkah

def analisa_kompleksitas():
    ukuran_data = [100, 1000, 10000]
    print(f"{'Ukuran':<10} | {'Waktu Sequential (s)':<22} | {'Waktu Binary (s)':<18}")
    print("-" * 60)

    for n in ukuran_data:
        data = list(range(n))
        target = n - 1

        start_seq = time.perf_counter()
        sequential_search_numeric(data, target)
        end_seq = time.perf_counter()
        t_seq = end_seq - start_seq

        start_bin = time.perf_counter()
        binary_search_numeric(data, target)
        end_bin = time.perf_counter()
        t_bin = end_bin - start_bin

        print(f"{n:<10} | {t_seq:<22.8f} | {t_bin:<18.8f}")

analisa_kompleksitas()
```

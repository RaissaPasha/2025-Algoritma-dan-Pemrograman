# Praktikum 04 – Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 07 – Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
      print("Sebelum fungsi dijalankan...")
      fungsi()
      print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

# Penugasan 3 Fungsi dengan Default Argument

```python
def salam(waktu="pagi", nama="Mahasiswa"):
    print(f"Selamat {waktu}, {nama}!")

# Panggil dengan argumen lengkap
salam("malam", "Budi")

# Panggil dengan sebagian argumen (hanya nama)
salam(nama="Siti")

# Panggil tanpa argumen
salam()
```

# Penugasan 4 *Fungsi dengan args

```python
def jumlahkan_semua(*angka):
    return sum(angka)

print(jumlahkan_semua(1, 2, 3))
print(jumlahkan_semua(1, 2, 3, 4, 5))
print(jumlahkan_semua(1, 2, 3, 4, 5, 6, 7))

```

# Penugasan 5 **Fungsi dengan kwargs

```python
def info_barang(**data):
    for k, v in data.items():
        print(f"{k}: {v}")

info_barang(Nama="Buku", Harga=15000, Stok=30)
```

# Penugasan 8 Fungsi Decorator

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, selamat datang!")

halo()
```

# Penugasan 9 Studi Kasus Nyata – Program Nilai

```python
def input_nilai():
    return [70, 80, 65, 90, 72]

def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

nilai_mahasiswa = input_nilai()
rata2 = hitung_rata(nilai_mahasiswa)
status = status_kelulusan(rata2)
akhir = nilai_akhir(75, 85)

print("Daftar nilai:", nilai_mahasiswa)
print("Rata-rata:", rata2)
print("Status kelulusan:", status)
print("Nilai akhir (bobot 40-60):", akhir)
```

# Praktikum 1 Buble Short

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

# Praktikum 2 Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

# Praktikum 3 Insertion Sort

```python
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

# Praktikum 4 Studi Kasus: Mengurutkan Nilai Mahasiswa


```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

# Praktikum 5 Perbandingan Waktu Eksekusi Sorting

```python
import random, time
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

data = [random.randint(1, 1000) for _ in range(500)]

for func in [bubble_sort, selection_sort, insertion_sort]:
  salinan = data.copy()
  start = time.time()
  func(salinan)
  end = time.time()
print(f"{func.__name__} membutuhkan waktu {end - start:.5f} detik")
```

# Praktikum 6 Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
        else:
          if data[j][key_index] > data[j + 1][key_index]:
              data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen",
10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(),
0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(),
1, reverse=True))
```

# Penugasan 5

```python
import random
import pandas as pd
import copy

def bubble_sort_count(arr):
    data = arr.copy()
    comparisons = 0
    n = len(data)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            comparisons += 1
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                swapped = True
        if not swapped:
            break
    return comparisons

def selection_sort_count(arr):
    data = arr.copy()
    comparisons = 0
    n = len(data)
    for i in range(n - 1, 0, -1):
        max_idx = 0
        for j in range(1, i + 1):
            comparisons += 1
            if data[j] > data[max_idx]:
                max_idx = j
        if max_idx != i:
            data[i], data[max_idx] = data[max_idx], data[i]
    return comparisons

def insertion_sort_count(arr):
    data = arr.copy()
    comparisons = 0
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            comparisons += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key
    return comparisons

input_sizes = [10, 50, 100]
results = []

random.seed(42)

for size in input_sizes:
    test_data = [random.randint(1, 1000) for _ in range(size)]

    b_count = bubble_sort_count(test_data)
    s_count = selection_sort_count(test_data)
    i_count = insertion_sort_count(test_data)

    results.append({
        "Input Size (N)": size,
        "Bubble Sort": b_count,
        "Selection Sort": s_count,
        "Insertion Sort": i_count
    })

df = pd.DataFrame(results)
print(df.to_string(index=False))
```

```python
def input_barang():
    barang_list = []
    n = int(input("Masukkan jumlah barang: "))
    for i in range(n):
        print(f"\nData barang ke-{i+1}")
        nama = input("Nama  : ")
        harga = float(input("Harga : "))
        barang_list.append({"nama": nama, "harga": harga})
    return barang_list

def tampilkan(barang_list):
    print("\nDaftar Barang:")
    for b in barang_list:
        print(f"- {b['nama']} : Rp {b['harga']}")

def urutkan_barang(barang_list):
    print("\n=== Pilih Kriteria Sortir ===")
    print("1. Nama")
    print("2. Harga")
    kriteria = input("Pilihan (1/2): ")

    print("\n=== Pilih Arah Sortir ===")
    print("1. Ascending (kecil → besar / A → Z)")
    print("2. Descending (besar → kecil / Z → A)")
    arah = input("Pilihan (1/2): ")

    if kriteria == "1":
        key_func = lambda x: x["nama"]
    elif kriteria == "2":
        key_func = lambda x: x["harga"]
    else:
        print("Pilihan kriteria tidak valid.")
        return barang_list

    reverse_flag = True if arah == "2" else False

    return sorted(barang_list, key=key_func, reverse=reverse_flag)

barang = input_barang()
tampilkan(barang)

barang = urutkan_barang(barang)
print("\nSetelah diurutkan:")
tampilkan(barang)
```

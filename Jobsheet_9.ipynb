# Penugasan Modul Buatan Sendiri

```python
%%writefile konversi.py

def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
import konversi

print(konversi.celcius_ke_fahrenheit(32))
print(konversi.fahrenheit_ke_celcius(56))
```

```python
%%writefile lap.py

def ringkasan_skor(skor, total):
    print("="*30)
    print("RINGKASAN HASIL")
    print(f"Skor Benar    : {skor}")
    print(f"Total Soal    : {total}")
    persen = (skor / total) * 100
    print(f"Persentase    : {persen:.2f} %")
    if persen >= 60:
        print("Status        : Lulus")
    else:
        print("Status        : Tidak Lulus")
    print("="*30)
```

```python
import random
import lap

def main():
    skor = 0
    total = 5

    for i in range(total):
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        print(f"Soal {i+1}: {a} + {b} = ?")
        try:
            jawab = int(input("Jawab: "))
            benar = jawab == (a + b)
        except ValueError:
            benar = False
        if benar:
            print("Benar!\n")
            skor += 1
        else:
            print(f"Salah. Jawaban yang benar: {a + b}\n")
    lap.ringkasan_skor(skor, total)

if __name__ == "__main__":
    main()
```
